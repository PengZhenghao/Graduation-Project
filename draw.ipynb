```python
import matplotlib.pyplot as plt
import matplotlib
import pickle
import numpy as np
from detector import DetectedObject
% matplotlib inline
```

```python
raw = pickle.load(open("data/0505-800m.pkl", "rb"))
data = [x["Object 0"]["centroid"] for x in raw if "Object 0"in x]
data = np.stack(data)
```

```python
plt.plot(data[:2700,0])
```

```python
pdata = data[:2710]
means = []
for i in range(2610):
    means.append(pdata[i:i+100].mean(axis=0))
means = np.stack(means)
```

```python
plt.figure(figsize=(10,6), dpi=300)
plt.plot(np.arange(50, 2660), means[:,0], c="C0", label="MA of x")

plt.plot(np.arange(50, 2660), means[:,1], c="C1", label="MA of y")

plt.plot(pdata[:,0], c="C0", label="x", alpha=0.5)

plt.plot(pdata[:,1], c="C1", label="y", alpha=0.5)

plt.legend()

plt.xlabel("Time Step")

plt.ylabel("Position (m)")
```
